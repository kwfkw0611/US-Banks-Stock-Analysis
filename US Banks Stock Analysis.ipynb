### Market analyis for US  banks from Jan 1, 2007 to Jan 1, 2018 


```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_datareader import data, wb
from datetime import datetime
%matplotlib inline
plt.rcParams["patch.force_edgecolor"] = True
pd.set_option('display.max_columns', 40)
```

### Bank data

Bank stock information:
    1. Bank of America (BAC)
    2. CitiGroup (C)
    3. Goldman Sachs (GS)
    4. JPMorgan Chase (JPM)
    5. Morgan Stanley (MS)
    6. Wells Fargo (WFC)

```python
# Market data from Jan 1, 2007 to Jan 1, 2018 are imported from Morningstar, Inc.
start = datetime(2007, 1, 1)
end = datetime(2018, 1, 1)
# Bank of America
BAC = data.DataReader('BAC', 'morningstar', start, end)

# CitiGroup
C = data.DataReader('C', 'morningstar', start, end)

# Goldman Sachs
GS = data.DataReader('GS', 'morningstar', start, end)

# JP Morgan
JPM = data.DataReader('JPM', 'morningstar', start, end)

# Morgan Stanley
MS = data.DataReader('MS', 'morningstar', start, end)

# Wells Fargo
WFC = data.DataReader('WFC', 'morningstar', start, end)


print(BAC.head())
print('*'*70)
print(C.head())
print("*"*70)
print(GS.head())
print("*"*70)
print(JPM.head())
print("*"*70)
print(MS.head())
print("*"*70)
print(WFC.head())
```

### Cleaning and combining bank data


```python
# dropping Symbol index and resetting index to Date
BAC.reset_index(level='Symbol',inplace=True)
BAC.drop('Symbol', axis=1, inplace=True)
BAC.head(10)
```

```python
C.reset_index(level='Symbol', inplace=True)
C.drop('Symbol', axis=1, inplace=True)
C.head()
```

```python
GS.reset_index(level='Symbol', inplace=True)
GS.drop('Symbol', axis=1, inplace=True)
GS.head()
```

```python
JPM.reset_index(level='Symbol', inplace=True)
JPM.drop('Symbol', axis=1, inplace=True)
JPM.head()
```

```python
MS.reset_index(level='Symbol', inplace=True)
MS.drop('Symbol', axis=1, inplace=True)
MS.head()
```

```python
WFC.reset_index(level='Symbol', inplace=True)
WFC.drop('Symbol', axis=1, inplace=True)
WFC.head()
```

```python
# concating individual market data to create one dataframe
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']

# concat column wise
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis= 1, keys = tickers)
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
bank_stocks.head()
```

```python
bank_stocks.info()
```

```python
# statistical summary can see 0 values that need to be dropped or filled
bank_stocks.describe()
```

```python
# dropping 0 values as they represent days when markets are closed(holidays)
def drop_zeroes(bank):
    for value in bank:
        ix_zeroes = bank[bank[value]== 0].index
    bank = bank.drop(ix_zeroes)
    return bank

clean_banks = drop_zeroes(bank_stocks)
clean_banks
```

```python
# searching for any missing data
col = list(clean_banks.columns[:5])
col = [i[1] for i in col]

for item in col:
    xs_bank = clean_banks.xs(key=item, axis=1, level='Stock Info').min()
    print(item, '\n', xs_bank,'\n')
```

```python
# Bank of American and Citigroup Volume missing data
clean_banks[(clean_banks[('BAC','Volume')] == 0) | (clean_banks[('C', 'Volume')] == 0)]
```

```python
# updating Volume missing data 
clean_banks = clean_banks.astype(str)
clean_banks.loc['2008-12-15', 'BAC']['Volume'] = clean_banks.loc['2008-12-15', 'BAC']['Volume'] = '81959400'
clean_banks.loc['2008-12-18', 'BAC']['Volume'] = clean_banks.loc['2008-12-18', 'BAC']['Volume'] = '102574800'
clean_banks.loc['2008-12-15', 'C']['Volume'] = clean_banks.loc['2008-12-15', 'C']['Volume'] = '10147200'
```

```python
clean_banks = clean_banks.astype(float)
```

```python
# creating Banks Return value in a new dataframe from Banks market Close
returns = pd.DataFrame()
for tick in tickers:
    returns[tick+' Return'] = clean_banks[tick]['Close'].pct_change()
returns.tail()
```
